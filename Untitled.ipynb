```python
import gzip

# Define file paths using raw strings to handle backslashes or special chars
coldFile = r"5_28 Data Collection/nodb906.gz"
humanFile = r"5_28 Data Collection/nodb906_p.gz"
testFile = r"5_28 Data Collection/nodb906_p_no.gz"
s_off_file = r"5_28 Data Collection/nodb906.gz"

# Function to load and parse data from a .gz file
def load_data_gz(file_path):
    data = []
    with gzip.open(file_path, 'rt') as f:  # 'rt' = read text mode
        for line in f:
            parts = line.strip().split(' ')
            parsed_line = [float(num.strip('() ')) for num in parts if num.strip()]
            data.append(parsed_line)
    return data

# Load data
dataCold = load_data_gz(coldFile)
dataHuman = load_data_gz(humanFile)
dataSoff = load_data_gz(s_off_file)
dataTest = load_data_gz(testFile)
```

```python
def chisq(x_obs, y_obs, func, params, err=None, err_p=None, xlim=None):
    
    
    y_exp = np.array(func(x_obs, *params))
    
    if xlim is not None:
        i1 = (np.abs(np.array(x_obs) - xlim[0])).argmin()
        i2 = (np.abs(np.array(x_obs) - xlim[1])).argmin()
        
        y_exp = y_exp[i1:i2]
        x_obs = x_obs[i1:i2]
        y_obs = y_obs[i1:i2]
    
    chisquare = sum((np.array(y_obs) - y_exp)**2/y_exp)
    chisq_nu = chisquare/(len(x_obs)- len(params))
    
    
    if err is not None:
        if xlim: err = err[i1:i2]
        y_exp = np.array(func(x_obs, *err_p))
        chisquare_w = sum((np.array(y_obs) - y_exp)**2/np.array(err)**2)
        chisq_nu_w = chisquare_w/(len(x_obs)- len(params))
        
        print(f"Chi_sq: {chisquare:.3f}\t Reduced chi_sq.: {chisq_nu:.3f}, # pts: {i2-i1}\nWeighted chi_sq: {chisquare_w:.3f}\t Reduced weighted chi_sq.: {chisq_nu_w:.3f}")
        
        return [(chisquare, chisq_nu), (chisquare_w, chisq_nu_w), i2-i1]
    
    print(f"Chi_sq: {chisquare:.3f}\t Reduced chi_sq.: {chisq_nu:.3f}, # pts: {i2-i1}")
        
    return [(chisquare, chisq_nu)]
        
    
```
