```python
import gzip

# Define file paths using raw strings to handle backslashes or special chars
coldFile = r"5_28 Data Collection/nodb906.gz"
humanFile = r"5_28 Data Collection/nodb906_p.gz" 
testFile = r"5_28 Data Collection/nodb906_p_no.gz"
s_off_file = r"5_28 Data Collection/nodb906.gz"
cold_low = r"5_28 Data Collection/nodb.gz"
human_low = r"5_28 Data Collection/nodb_p.gz" 

# Function to load and parse data from a .gz file
def load_data_gz(file_path):
    data = []
    with gzip.open(file_path, 'rt') as f:  # 'rt' = read text mode
        for line in f:
            parts = line.strip().split(' ')
            parsed_line = [float(num.strip('() ')) for num in parts if num.strip()]
            data.append(parsed_line)
    return data

# Load data
dataCold = load_data_gz(coldFile)
dataHuman = load_data_gz(humanFile)
dataSoff = load_data_gz(s_off_file)
dataTest = load_data_gz(testFile)
dataCold_low = load_data_gz(cold_low)
dataHuman_low = load_data_gz(human_low)
```

```python
import scipy.stats as sps
import scipy.optimize as spo
import scipy as sp
from scipy.optimize import curve_fit
from scipy.stats import norm

data2 = np.mean(np.real( (dataCold_low + dataCold_high)/2  )*1e3, axis = 0) #mV

bins = 40
fig, ax = plt.subplots(figsize = (10, 8))
ax.hist(data2, bins = bins,histtype = 'step', color = 'blue', density = False)

ax.set_title("Radio-Telescope Pointing at Zenith", fontsize = 17)
ax.set_xlabel("Voltage (mV)", fontsize = 15)
ax.set_ylabel("Number of Events", fontsize = 15)

mean = np.mean(data2)
std = np.std(data2)

#error in error = 1 / sqrt(2N - 2) ------ FOR STD gaussian distribution (we can assume this)
N = len(data2)
std_err = 1/(np.sqrt(2*N - 2))

#error in mean: (std of distribution) / sqrt(N) 

mean_err = std/(np.sqrt(N))

# print("Standard Deviation: {0:.3f} ± {1:.3f}".format(std, std_err))
# print("Mean: {0:.3f} ± {1:.3f}".format(mean, mean_err))

# ax.text(2, 125, f"Standard Dev.: {round(std, 3)} ± {round(std_err, 3)}", fontsize = 15)
# ax.text(2, 115, f"Mean: {round(mean, 3)} ± {round(mean_err, 3)}", fontsize = 15)


#Part B
y, bin_edges = np.histogram(data2, bins=bins, density = False)
bin_centers = 0.5*(bin_edges[1:] + bin_edges[:-1])

#error - poisson can be assumed to be sqrt(num_events_in_bin) where n = 1 for no events
poisson_err = []
for i in range(len(y)):
    if y[i] > 0:
        poisson_err.append(np.sqrt(y[i]))
    else: 
        poisson_err.append(1)

#ax.errorbar(bin_centers, y, yerr = poisson_err,marker = '.', mfc='red', mec = "black", ms=14,ecolor = "red", linestyle='')
ax.errorbar(bin_centers, y, yerr = poisson_err, fmt = "o", capsize = 3, color = 'red', label = "Error")


def gauss0(x, a, mean, std):
    return a* sps.norm.pdf(x, mean, std)

bin_hights, binedges, _ = ax.hist(data2, bins = bins, density = False, histtype = 'step', color = "blue", label = "Data Samples")

bin_centers = 0.5*(binedges[1:] + binedges[:-1])
x = np.linspace(min(data2), max(data2), len(data2))

params, cov = curve_fit(gauss0, bin_centers, bin_hights)

params_err, cov_err = curve_fit(gauss0, bin_centers, bin_hights, sigma = poisson_err)

y_binned = gauss0(x, *params)
y_binnederr = gauss0(x, *params_err)
ax.plot(x, y_binned, color = "green", label = "Gaussian Fit")

ax.plot(x, y_binnederr, color = "purple", ls = 'dashed', label = "Weighted Gaussian Fit")


ax.grid(True, axis = 'y')
ax.legend(fontsize = 16)

print(f"Mean: {mean:.4f} \pm {mean_err:.4f}")
print(f"Std: {std} \pm {std_err}")

print("fit")
print(f"Mean: {params[1]:.4f} \pm {np.sqrt(cov[1][1]):.4f}")
print(f"Std: {params[2]:.4f} \pm {np.sqrt(cov[2][2]):.4f}")

print("weighted fit")
print(f"Mean: {params_err[1]:.4f} \pm {np.sqrt(cov_err[1][1]):.4f}")
print(f"Std: {params_err[2]:.4f} \pm {np.sqrt(cov_err[2][2]):.4f}")

print()
#chi-2

# chi sqr 
#print(len(data2), len(y_binned), len(poisson_err))
observed = np.array(bin_hights)
expected = np.array(gauss0(bin_centers, *params))
chi2_weighted = sum(((observed-expected))**2/np.array(poisson_err)**2)
chi2 = sum((observed-expected)**2/expected)
reduced_chi2 = chi2/(bins - 3) 
reduced_chi2_weighted = chi2_weighted/(bins - 3)

print(chi2, reduced_chi2)
print(f"chi2: {chi2:.4f} , rchi2: {reduced_chi2:.4f}")
print(f"Wchi2: {chi2_weighted:.4f} , Wrchi2: {reduced_chi2_weighted:.4f}")
```
